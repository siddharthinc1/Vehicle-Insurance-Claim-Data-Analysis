```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("D:\Data Analyst\Projects\Vehicle Insurance Claim Data Analysis\insurance_claims.xlsx")
df
```

```python
df_plot = df.groupby(['Incident_city','Incident_type']).size().reset_index().pivot(columns='Incident_type', index='Incident_city', values=0)
df_plot
```

```python
df_plot.plot(kind='bar', stacked=True, ylabel='count', xlabel='Incident city', title='Distribution of Incident types within each City')
plt.show()
```

```python
sb.countplot(df, y='Incident_type')
plt.title('Counts of Observations in each Incident Type')
plt.ylabel('Incident types')
plt.xlabel('Count')
plt.show()
```

```python
df.groupby('Incident_severity').size().sort_values(ascending=False).plot(kind='pie', autopct='%1.0f%%')
plt.title('Percentage of each Incident Severities')
plt.show()
```

```python
sb.displot(df, x='Age', hue='Insured_gender', col='Insured_gender')
plt.show()
```

```python
sb.relplot(df, x='Age', y='Total_claim_amount', hue='Insured_gender', col='Insured_gender', palette=['b','r'])
plt.show()
```

```python
sb.stripplot(df, x='Injury_claim', y='Insured_gender', palette=['b','r'])
plt.title('Injury Claim According to Male vs Female')
plt.xlabel('Injury claim')
plt.ylabel('Genders')
plt.show()
```

```python
sb.stripplot(df, x='Property_claim', y='Insured_gender', palette=['b','r'])
plt.title('Property Claim According to Male vs Female')
plt.xlabel('Property claim')
plt.ylabel('Genders')
plt.show()
```

```python
sb.stripplot(df, x='Vehicle_claim', y='Insured_gender', palette=['b','r'])
plt.title('Vehicle Claim According to Male vs Female')
plt.xlabel('Vehicle claim')
plt.ylabel('Genders')
plt.show()
```
